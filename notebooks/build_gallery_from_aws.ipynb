```python
# to download the gallery from s3 
def downloadDirectoryFroms3(bucket, person_id_str, local_repo='./data/galleries/'):
    """" 
    bucket : s3 bucket i.e. a boto3 resource : boto3.resource('s3')
    person_id_str : string of the id of a person
    local_repo : local repo. to save the gallery 

    Load the gallery of the personnality and copy it in a local repo 

    """
    save_dir = local_repo + person_id_str + '/'
    for obj in bucket.objects.all():
        if id_str in obj.key:
            if not os.path.exists(os.path.dirname(save_dir)):
                os.makedirs(os.path.dirname(save_dir))
            tmp = str(obj.key)
            save_dir = local_repo + tmp[tmp.find(person_id_str):]
            #print(save_dir)
            bucket.download_file(obj.key, save_dir)
```

```python
# save the gallery in the local repo
for id_name in id_person:
    id_str = str(id_name)
    downloadDirectoryFroms3(bucket_root,id_str,'../data/galleries/')
```
